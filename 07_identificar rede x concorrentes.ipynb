```python
import pandas as pd
```

```python
path_enem = 'D:/000000001_ENEM/DADOS_PROCESSADOS/enem.csv'
path_unidades = 'D:/000000001_ENEM/DADOS_REDE/UNIDADES_REDE.xlsx'
path_concorrentes = 'D:/000000001_ENEM/DADOS_REDE/CONCORRENTES.xlsx'
path_out = 'D:/000000001_ENEM/DADOS_PROCESSADOS/'
```

```python
enem = pd.read_csv(path_enem, sep = ';', encoding = 'ISO-8859-1')
```

```python
enem.head()
```

```python
unidades_rede = pd.read_excel(path_unidades, encoding = 'ISO-8859-1')
unidades_concorrentes = pd.read_excel(path_concorrentes)
```

```python
unidades_rede
```

```python
unidades_rede['ind_rede'] = 1
```

```python
enem_unidades = pd.merge(enem, unidades_rede, left_on = 'CO_ESCOLA', right_on = 'co_entidade', how = 'left')
```

```python
enem_unidades.head()
```

```python
enem_unidades['ind_rede'].fillna(0, inplace = True)
```

```python
enem_unidades.head()
```

```python
enem_unidades.groupby(['ind_rede']).count()
```

```python
unidades_concorrentes ['ind_concorrente'] = 1
```

```python
unidades_concorrentes
```

```python
enem_completo = pd.merge(enem_unidades, unidades_concorrentes, left_on = 'CO_ESCOLA', right_on = 'co_entidade', how = 'left')
```

```python
enem_completo['co_entidade_y'].fillna(0)
```

```python
enem_completo.head()
```

```python
enem_completo[enem_completo['ind_concorrente']==1]
```

```python
enem_completo.rename(columns={'unidade_rede':'nome_entidade_rede','co_entidade_x':'cod_entidade_rede', 'no_entidade':'nome_concorrente','co_entidade_y':'cod_concorrente','unidade_concorrente': 'cod_entidade_concorrente'}, inplace=True)
```

```python
enem_completo['nome_concorrente'].fillna(0, inplace = True)
enem_completo['cod_concorrente'].fillna(0, inplace = True)
enem_completo['cod_entidade_concorrente'].fillna(0, inplace = True)
enem_completo['nome_entidade_rede'].fillna(0, inplace = True)
enem_completo['cod_entidade_rede'].fillna(0, inplace = True)
```

```python
enem_completo['chave'] = (enem_completo['NU_ANO'].astype(int)).astype(str) +  '|' +  (enem_completo['CO_ESCOLA'].astype(int)).astype(str)
```

```python
enem_completo['chave']
```

```python
enem_completo.to_csv(path_out+'enem_completo.csv', sep = ';', index = None, encoding = 'ISO-8859-1')
```
