```python
import pandas as pd
```

```python
co_ufs = [43, 53, 51]
## 43 - RS
## 51 - MT
## 53 - DF
path_escolas = 'D:/000000001_ENEM/ESCOLAS/'
```

```python
escolas_2019 = pd.read_csv(path_escolas+'ESCOLAS_2019.csv', encoding = 'ISO-8859-1', sep = '|', low_memory=False)
for uf in co_ufs:
    path = path_escolas+'escolas_2019_{}.csv'.format(uf)
    escolas_2019[escolas_2019['CO_UF'] == uf].to_csv(path, sep = ';', index = None, encoding = 'ISO-8859-1')
```

```python
escolas_2018 = pd.read_csv(path_escolas+'ESCOLAS_2018.csv', encoding = 'ISO-8859-1', sep = '|', low_memory=False)
for uf in co_ufs:
    path = path_escolas+'escolas_2018_{}.csv'.format(uf)
    escolas_2018[escolas_2018['CO_UF'] == uf].to_csv(path, sep = ';', index = None, encoding = 'ISO-8859-1')
```

```python
escolas_2017 = pd.read_csv(path_escolas+'ESCOLAS_2017.csv', encoding = 'ISO-8859-1', sep = '|', low_memory=False)
for uf in co_ufs:
    path = path_escolas+'escolas_2017_{}.csv'.format(uf)
    escolas_2017[escolas_2017['CO_UF'] == uf].to_csv(path, sep = ';', index = None, encoding = 'ISO-8859-1')
```

```python
escolas_2016 = pd.read_csv(path_escolas+'ESCOLAS_2016.csv', encoding = 'ISO-8859-1', sep = '|', low_memory=False)
for uf in co_ufs:
    path = path_escolas+'escolas_2016_{}.csv'.format(uf)
    escolas_2016[escolas_2016['CO_UF'] == uf].to_csv(path, sep = ';', index = None, encoding = 'ISO-8859-1')
```

```python
escolas_2015 = pd.read_csv(path_escolas+'ESCOLAS_2015.csv', encoding = 'ISO-8859-1', sep = '|', low_memory=False)
for uf in co_ufs:
    path = path_escolas+'escolas_2015_{}.csv'.format(uf)
    escolas_2015[escolas_2015['CO_UF'] == uf].to_csv(path, sep = ';', index = None, encoding = 'ISO-8859-1')
```

```python
geral_escolas = pd.concat([escolas_2019, escolas_2018, escolas_2017, escolas_2016, escolas_2015], sort = False)
```

```python
geral_escolas['chave'] = geral_escolas['NU_ANO_CENSO'].astype(str) +'|'+  geral_escolas['CO_ENTIDADE'].astype(str)
```
